# using pandas one hot encos

```python
pd.get_dummies(data=df, columns=['drivewheel','CarName','doornumber','fueltype'], dtype=int, drop_first='first')
```

# with scikit learn one hot encoder 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', dtype=int, sparse_output=False)
```

```python
df_trf = ohe.fit_transform(df[['fueltype','drivewheel','CarName','doornumber']])
```

```python
df_trf
```

```python
cols = ohe.get_feature_names_out()
```

```python
cols
```

```python
new_df = pd.DataFrame(data=df_trf, columns=cols)
```
